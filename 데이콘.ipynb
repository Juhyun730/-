```python
val_df = pd.read_csv('/content/drive/MyDrive/dacon_신카/val.csv') # Validation
val_df.head()
```

```python
val_normal, val_fraud = val_df['Class'].value_counts() #밸리데이션 0,1 개수 세기
val_contamination = val_fraud / val_normal
print(f'Validation contamination : [{val_contamination}]')
```

```python
print(val_normal, val_fraud)
```

```python
# Train dataset은 Label이 존재하지 않음
train_x = train_df.drop(columns=['ID']) # Input Data
```

Isolation Forest는 Unsupervised Anomaly Detection 중 하나로 현재 갖고 있는 데이터 중 이상치를 탐지할 때 주로 사용된다. 여러 개의 의사결정나무 (decision tree)를 종합한 앙상블 기반의 이상탐지 기법으로 의사결정나무를 지속적으로 분기시키면서 모든 데이터 관측치의 고립 정도 여부에 따라 이상치를 판별하는 방법이다.

비정상데이터는 의사결정 나무의 루트에서 가까운 깊이에서 고립되고, 정상 데이터는 루트에서 먼 깊이에서 고립되는 방식이다.
즉 특정한 샘플이 고립되는 leaf노드까지의 거리를 outlier score로 정의하고 루트 노드까지의 평균거리가 짧을 수록 outlier score가 높아지는 원리이다.

```python
# Train dataset도 Validation dataset과 동일한 비율로 사기거래가 발생 했을 것이다라고 가정하기
model = IsolationForest(n_estimators=125, max_samples=len(train_x), contamination=val_contamination, random_state=42, verbose=0)
model.fit(train_x)
```

```python
def get_pred_label(model_pred):
    # IsolationForest 모델 출력 (1:정상, -1:불량(사기)) 이므로 (0:정상, 1:불량(사기))로 Label 변환
    model_pred = np.where(model_pred == 1, 0, model_pred)
    model_pred = np.where(model_pred == -1, 1, model_pred)
    return model_pred
```

```python
val_x = val_df.drop(columns=['ID', 'Class']) #x값
val_y = val_df['Class'] 

val_pred = model.predict(val_x) # 모델 예측
val_pred = get_pred_label(val_pred)
val_score = f1_score(val_y, val_pred, average='macro') #모델의 f1스코어: 마크로f1스코어를 사용하기
print(f'Validation F1 Score : [{val_score}]')
print(classification_report(val_y, val_pred))
```
